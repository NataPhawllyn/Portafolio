```python
print("------------------------------ PUNTO 3--------------------------------")

#A)
def factorial(n):
    """
    esta  funcion calcula el factorial n!
    Entrada----
    n : [int] numero entero
    Salida----
    Fac_n :[int] resultado del factorian n!

    """
    Fac_n=1
    if n!=0 or n!=1:
      for i in range(1,n+1):                                                    # cambie  la recursividad por un for por que  me estaba  dando errores  de profundidad
        Fac_n=Fac_n*i
    return Fac_n

def binomial(n,k):
    """
    esta  funcion calcula el coeficientes binomia lpara dos numeros dados
    Entrada----
    n : [int] numero entero positivo mayor que K
    k : [int] numero entero positivo mayor o igual a 1
    Salida----
    bin_nk :[int] coeficiente binomial para n, k

    """

    b=n-k
    bin_nk =factorial(n)//(factorial(k)*factorial(b))
    return int(bin_nk)

## prueba de la funcion
# asignamos valores para n y k ,enteros positivos, con n mayor que  k
n=5
k=3
print("El coeficiente binomial para n={}, k={} es {}".format(n,k,binomial(n,k)))


#B)
# triangulo de Pascal

for n in range(21):                                                       # itero entre las filas  imprimiendo por  fila
     print(*(f"{binomial(n, k):<10}" for k in range(n+1)))                  # Imprimir coeficientes binomiales para cada fila

#C)
 #I)

def probabilidad(n,k):
    """
    esta  funcion calcula la probabilidad  de una  moneda  lanzada  n veces salga  k  caras
    Entrada----
    n : [int] numero  de lanzamientos
    k : [int] numero  de caras
    Salida----
    probbilidad :[float] probabilidad

    """
    return binomial(n,k)/2**n

print("\n La  probabilidad de que salga  {} caras  en {} lanzamientos  es {}".format(100,60,binomial(100,60)))

#II)
def probabilidad_mas_K(n,k):
    """
    Esta función calcula la probabilidad de que salgan  k caras o más en n lanzamientos.
    Entrada----
    n : [int] número de lanzamientos
    k : [int]  numero de  caras
    Salida----
    probabilidad : [float] probabilidad
    """
    prob_total = 0
    for i in range(k, n + 1):
        prob_total += probabilidad(n, i)                                # sumo  todas las  probabilidades despues de k
    return prob_total

print("\nLa probabilidad de que salgan {} caras o más en {} lanzamientos es {}".format(60,100, probabilidad_mas_K(100,60)))

```
