```python
import  pandas as pd
import matplotlib.pyplot as plt
import numpy as np

arr = pd.read_csv("transistors-per-microprocessor.csv")

def moore(y1,y0=arr["Year"][0],n0=arr["Transistors per microprocessor"][0],T2=2):
  return(n0*2**((y1-y0)/T2))

#display(arr)
arr.head()

x=arr["Year"]
arr["moore"]=moore(arr["Year"])

plt.plot(arr["Year"],np.log(arr["Transistors per microprocessor"]),label="datos")
plt.plot(arr["Year"],np.log(arr["moore"]),marker="*",label="Moore's Law")
plt.xlabel("a√±os")
plt.ylabel("num, Trancistores")
plt.legend()
```
