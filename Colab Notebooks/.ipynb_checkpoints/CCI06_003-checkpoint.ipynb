```python
year = 2024
if (year % 4 == 0) and (year % 100!=0):
  is_leap_year = True
elif (year % 100 == 0) and (year % 400 == 0):
  is_leap_year = True
else:
  is_leap_year = False
ans = "is a" if is_leap_year else "is not a"
print("{} {:s} leap year".format(year, ans))
```
