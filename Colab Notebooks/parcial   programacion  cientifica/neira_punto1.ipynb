```python
# Tarea 1 Nataly Phawllyn Neira Parra Cod. 614212782
# ComputaciÃ³n Cientific I 2024-I

print("------------------------- PUNTO 1-------------------------------------")
#asignamos los valores de las constantes
masa_e=9.1094E-31
carga_e=1.602E-19
epsilon=8.854E-12
cte_h=6.626E-34

def nivel_energia(n):
    """
    esta  funcion calcula el n- esimo nivel de energia
    Entrada----
    n : [int] nivel de energia
    Salida----
    E_n :[float] la  energia en dicho nivel

    """
    cte=-(masa_e*(carga_e**4))/(8*(cte_h*epsilon)**2)                           #calulamos una constante apartir de las constantes
    E_n = cte/(n**2)
    return E_n

#A)
# se imprimeusando  el formato  e iterando dentro del prin con un salto delinea en cada iteracin
print(*(f"El nivel de energia para el {n} nivel es {nivel_energia(n)}" for n in range(1, 21)), sep='\n')

#B)

import sys # libreria para exportar el .txt

def diferencia_energia(n_i,n_f):
    """
    esta  funciÃ³n calcula ladiferencia  de energia entre dos niveles
    Entrada----
    n_0 : [int] nivel de energia inicial
    n_f : [int] nivel de energia final
    Salida----
    dif_energia :[float] diferencia denergia entre los niveles dados

    """
    dif_energia= nivel_energia(n_i)-nivel_energia(n_f)
    return dif_energia


 # este codigo es un gestor de texto que  genera el .txt desde un print
with open('dif_energia.txt', 'w') as file:
    sys.stdout = file                                                           # Redirige la salida estandar al archivo
    for i in range(1, 6):                                                       # itero entre las filas  imprimiendo por  fila
        print(*(f"{diferencia_energia(i,f):<30}" for f in range(1, 6)))         #imprimo sin salto de linea  para las columnas agregando 30 espacios fijos entre  impresion
    sys.stdout = sys.__stdout__                                                 # Restaura la salida estandar
```
