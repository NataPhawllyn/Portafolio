```python
print("------------------------------ PUNTO 5--------------------------------")
import math
#A)
def trapezint1(f, a, b):
    """
    Esta función calcula la aproximación de la integral definida utilizando la regla del trapecio.

    Entrada----
    f : [función] función a integrar
    a : [float] límite inferior de la integral
    b : [float] límite superior de la integral
    Salida----
    aproximacion : [float] aproximación del valor de la integral
    """
    aproximacion = (b - a) / 2 * (f(a) + f(b))
    return aproximacion

# defino la  función f(x)
def f(x):return math.sin(x)


#B)
#  definimos  los limites de la  integral
a = 0
b = math.pi/2

#Calculo el error

error = abs(1 -trapezint1(f, a, b))



print(" Para la integral del enunciado  tenemos  :\nValor exacto: {}\nAproximación: {}\nError: {}".format(1, trapezint1(f, a, b), error))
```
