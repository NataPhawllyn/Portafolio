```python
print("--------------------------- PUNTO 2-----------------------------------")

#A)

def factorial(n):
    """
    esta  funcion calcula el factorial n!
    Entrada----
    n : [int] numero entero
    Salida----
    Fac_n :[int] resultado del factorian n!

    """
    if n==0 or n==1 : Fac_n=1
    else:
        Fac_n=n*factorial(n-1)
    return Fac_n

print("el valor para {}! es {}".format(4,factorial(4)))

#B)
import random  #libreria que genera nÂœmeros aleatorios
import math


def taylor_sen(x,N):
    """
    esta  funcion calcula la aproximacion del sen(x) usando series de Taylor
    Entrada----
    x : [int] numero entero positivo
    N : [int] numero de terminos de la serie
    Salida----
    sen_aprox :[float] aproxximacion de sen(x)

    """
    Sen_x = 0

    for n in range(N):                                                          #para  facilitar la escritura de la formula,
        impar = 2*n+1                                                           # calulamos sus factores por aparte
        signo = (-1) ** n
        pot_x = x**impar
        fac = factorial(impar)
        n_termino= (signo * (pot_x / fac))                                     # calculo del termino n-esimo
        Sen_x+= n_termino                                                       # calculo sumatoria
        #print(impar, signo, pot_x, fac, n_termino, Sen_x)   #test de valores

    return Sen_x


# prueba de la funcion
# creamos un numero entero positivo  aleatorio  entre 0 y 360 y lopasamos a radianes
x=math.radians(random.randint(0, 360))
print("El valor aproximado del sen({}) es {} ".format(x,taylor_sen(x,25)))
```
