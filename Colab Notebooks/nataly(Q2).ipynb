```python
import numpy as  np
import math
import matplotlib.pyplot as plt
```

```python
def gaussian(delta,x):
  g=np.zeros(len(x))
  g=(1/(delta*np.sqrt(2*np.pi)))*np.exp(-x**2/(2*delta**2))
  return g
x=np.linspace(-10,10,100)


plt.plot(x,gaussian(1,x))
plt.plot(x,gaussian(1.5,x))
plt.plot(x,gaussian(2,x))
plt.labe
```
