```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.txt", delimiter=' ',header=None)
data
```

```python
X=data[1].values # variable  descriptiva
Y=data[2].values # variable  dependiente

X=X.reshape(-1, 1)
Y=Y.reshape(-1, 1)

c = np.linalg.inv(X.T @ X) @ X.T @ Y
print("La regresión lineal  y(x)=cx  que  mejor  ajusta  a los datos esta dada para  c=",c)
```

```python
X1=np.linspace(min(X),max(X),10)
Y1=c*X1

plt.figure(figsize=(10, 6))
sns.scatterplot(x=data[1], y=data[2], label="Datos")
plt.plot(X1, Y1, color='red', label="Regresión Lineal")
plt.legend()
plt.title('Relación Area vs  Costo de un terreno ')
plt.xlabel('Area [Acres]')
plt.ylabel('Costo [millones de dollares]')
plt.show()
```

```python
X_pred=np.array([25, 30, 35, 40])
Y_pred=c[0]*X_pred

for  i in range(4):
  print("La  predición  del  costo  para {}  acres  es de {:.2f} millones de dolares.".format(X_pred[i], Y_pred[i]))
```
