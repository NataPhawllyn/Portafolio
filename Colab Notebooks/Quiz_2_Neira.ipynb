```python
import numpy as np

def cadena_Collatz(n):
  lista=np.array([])
  lista=np.append(lista,n)
  while (n>0)&(n!=1):
    if n%2==0:
      n=n/2
    else:
      n=3*n+1
    lista=np.append(lista,n)
  return len(lista)

numeros=np.linspace(1,1000000,1000000)
largo_cadenas=np.array([])
for i  in numeros:
  largo_cadenas=np.append(largo_cadenas,cadena_Collatz(i))
mayor=np.argmax(largo_cadenas)
print("El nuúmero  con la  cadena mas  larga es ",numeros[mayor])
```
